```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 30, 500)
y = np.sin(x) * np.exp(-x/10)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, lw=2)
ax.axis('tight')
ax.set_title("axis('tight')");
```
