```python
%matplotlib inline
import matplotlib.pyplot as plt

days = [10,11,12,13,14,15,16]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.subplot(2,3,2)
plt.plot(days, max_t, linestyle='--', color='k', marker='o', label='max')
plt.legend(loc='lower left', fontsize='large', shadow=True)
plt.grid(True)
plt.title('Seoul Max Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')

plt.subplot(2,3,3)
plt.plot(days, avg_t, color='#2714e7', label='average')
plt.legend(loc='lower left', fontsize='large', shadow=True)
plt.grid(True)
plt.title('Seoul Average Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')

plt.subplot(2,3,4)
plt.plot(days, min_t, linestyle='--', color='k', marker='o', label='min')
plt.legend(loc='lower left', fontsize='large', shadow=True)
plt.grid(True)
plt.title('Seoul Min Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')

plt.tight_layout()

plt.savefig('plot.png')

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

days = [10,11,12,13,14,15,16]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6))  = plt.subplots(nrows=2, ncols=3, sharex=True)

ax1.plot(days, max_t, linestyle='--', color='k', marker='o', label='max')
ax1.legend(loc='lower left', fontsize='large', shadow=True)
ax1.grid(True)
ax1.set_title('Seoul')
#ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature')

ax5.plot(days, avg_t, color='#2714e7', label='average')
ax5.legend(loc='lower left', fontsize='large', shadow=True)
ax5.grid(True)
#ax5.set_title('Seoul Average Temperature')
#ax5.set_xlabel('Date')
ax5.set_ylabel('Temperature')

ax3.plot(days, min_t, linestyle='--', color='k', marker='o', label='min')
ax3.legend(loc='lower left', fontsize='large', shadow=True)
ax3.grid(True)
#ax3.set_title('Seoul Min Temperature')
ax3.set_xlabel('Date')
ax3.set_ylabel('Temperature')

plt.tight_layout()

plt.show()

fig.savefig('plot.png')
```
