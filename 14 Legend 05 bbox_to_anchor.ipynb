```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)
```

```python
fig, ax = plt.subplots(figsize=(6.7,3))

ax.set_xlim(-1, 1)
ax.set_ylim(-8, 8)
ax.set_xticks([-1.0,-0.5,0.0,0.5,1.0])
ax.set_yticks([-8,-6,-4,-2,0,2,4,6,8])

for beta in range(1,9):
    ax.plot(x, beta*x, label=f"y(x)={beta}*x")
ax.legend(ncol=4, loc=3, bbox_to_anchor=(0, 1));
```
