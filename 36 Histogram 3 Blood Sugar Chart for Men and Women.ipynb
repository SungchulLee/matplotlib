```python
import matplotlib.pyplot as plt
```

```python
x1 = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
x2 = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
fig, ax = plt.subplots()
ax.hist([x1, x2], 
        bins=[40,60,80,100,120,140,160,180], 
        rwidth=0.8,
        orientation='horizontal',
        label=['men','women'])
ax.legend(loc='upper right', shadow=True, fontsize='large')   
ax.set_xlabel("Sugar Level")
ax.set_ylabel("Number Of Patients")
ax.set_title("Blood Sugar Chart")
```
