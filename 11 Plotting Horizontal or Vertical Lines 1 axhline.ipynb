We flip a fair coin and record 1 if head, 0 otherwise. In fact we flip a fair coin many times and record the sample mean as a function of the number of flips.

```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
import pandas as pd
```

```python
# parameters of B(n,p)
n = 1
p = 0.5
```

```python
# number of simulation
n_sim = 1000
```

```python
coin = np.random.binomial(n, p, (n_sim,))
coin_cum_avg = np.cumsum(coin)/(np.arange(n_sim)+1)
```

```python
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(coin_cum_avg)
ax.axhline(y=p, ls='--', c='r');
```

```python
df = pd.DataFrame(coin.reshape((-1, 1)))
```

```python
fig, ax = plt.subplots(figsize=(12, 3))
df[0].expanding().mean().plot(ax=ax)
ax.axhline(y=p, ls='--', c='r');
```
