```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
```

```python
# number of data
N = 50

# 4-dim data
x = np.random.uniform(0., 1., (N,))
y = np.random.uniform(0., 1., (N,))
z = np.random.uniform(0., 1., (N,))
info = 1000 * np.random.uniform(0., 1., (N,))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=info, alpha=0.5)
#ax.plot(x, y, '.k')
```
