```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,1000)
y = np.sqrt(1 - x**2)
y2 = -y
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='b')
ax.plot(x, -y, color='b')
ax.axis('equal');
```
