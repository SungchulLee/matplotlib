```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x = np.random.multivariate_normal(mean, cov, 10000)
print(x.shape)
```

```python
fig, ax = plt.subplots()
a = ax.hexbin(x[:,0], x[:,1], gridsize=30, cmap='Blues')
plt.colorbar(a, label='counts in bin')
```
