```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
y1 = x
y2 = x + x**2
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, color='b', label=f"y(x)=x")
ax.plot(x, y2, color='g', label=f"y(x)=x+x**2")
ax.legend(fontsize=20);
```
