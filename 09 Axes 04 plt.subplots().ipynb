```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,3)
y = x ** 2
y2 = np.sqrt(x)
```

```python
if 0:
    fig, ax = plt.subplots()
    ax.plot(x, y);
elif 0:
    fig, axes = plt.subplots(1, 2, figsize=(12, 3))
    axes[0].plot(x, y)
    axes[1].plot(x, y2)
elif 0:
    fig, (ax, ax2) = plt.subplots(2, 1)
    ax.plot(x, y)
    ax2.plot(x, y2)
elif 0:
    fig, axes = plt.subplots(2, 3, figsize=(12, 3))
    axes[0,0].plot(x, y)
    axes[0,1].plot(y, x)
    axes[0,2].plot(x, y)
    axes[1,0].plot(x, y2)
    axes[1,1].plot(y2, x) 
    axes[1,2].plot(x, y2)
elif 1:
    fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2, 3, figsize=(12, 4))
    ax1.plot(x, y)
    ax2.plot(y, x)
    ax3.plot(x, y)
    ax4.plot(x, y2)
    ax5.plot(y2, x)
    ax6.plot(x, y2)
    fig.tight_layout() 
```
