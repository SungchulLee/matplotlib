```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1)
y1 = x ** 2
y2 = np.sqrt(x)
```

```python
fig = plt.figure(figsize=(12, 6))

for s in range(11):
    ax = fig.add_subplot(3, 4, s + 1)
    if s%2 == 0:
        ax.plot(y1, label="state {}".format(str(s)))
    else:
        ax.plot(y2, label="state {}".format(str(s)));
```
