```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x = np.linspace(-4, 4, 20)
y = np.linspace(-4, 4, 20)
X, Y = np.meshgrid(x, y)
Z = X + Y
```

```python
fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  
ax.plot_surface(X, Y, Z,
                rstride=2,
                cstride=2,
                cmap='viridis',
                linewidth=0.5,
                antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z=x+y');
```
