```python
days = [0,1,2,3,4,5,6]
max_t = [30,31,32,28,27,29,26]
min_t = [13,12,10,14,13,15,17]
avg_t = [25,28,28,26,20,22,21]
```

```python
import matplotlib.pyplot as plt

plt.plot(days, max_t, c='r', ls='-.', marker='+', label="max")
plt.plot(days, avg_t, c='b', ls='-', marker='o', label="avg")
plt.plot(days, min_t, c='g', ls='--', marker='h', label="min")

plt.legend(loc='upper right', shadow=True, fontsize='large')

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.grid()

plt.show()
```
