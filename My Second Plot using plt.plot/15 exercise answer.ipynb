1. Plot $y_i=1+x_i+\varepsilon_i$ as a scatter plot
where $x_i\sim N(0,1^2)$,
$\varepsilon_j\sim N(0,0.5^2)$,
and they are independent.
$\{(x_i,y_i)\}$ is a dataset. 
Plot $y=1+x$ as a line plot.
$y=1+x$ is an ideal regression line.

<div>
<img src="img/linear_regression.png" width="400"/>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)

n = 100

x = np.random.normal(size=(n,))

noize_level = 0.5
noize = noize_level * np.random.normal(size=(n,)) 

y = 1 + x + noize 
y_pred = 1 + x

plt.plot(x,y,'o')
plt.plot(x,y_pred)

plt.savefig('img/linear_regression.png')

plt.show()
```
