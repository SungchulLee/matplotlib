```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def normal_pdf(x):
    return np.exp(-x**2/2) / np.sqrt(2 * np.pi)
```

***ymin : scalar, optional, default: 0***

    Should be between 0 and 1, 0 being the bottom of the plot, 1 the top of the plot.
    
***ymax : scalar, optional, default: 1***

    Should be between 0 and 1, 0 being the bottom of the plot, 1 the top of the plot.

[DOC](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.axes.Axes.axvline.html)

```python
x = np.linspace(-3, 3)
y = normal_pdf(x)

fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(x, y)
ax.set_ylim(0, 0.5)
ax.axvline(x=-1.96, ymin=0, ymax=normal_pdf(-1.96)/0.5, c='r')
ax.axvline(x=1.96, ymin=0, ymax=normal_pdf(1.96)/0.5, c='r');
```
