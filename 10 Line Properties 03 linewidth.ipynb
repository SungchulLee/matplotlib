```python
import matplotlib.pyplot as plt
import numpy as np
import sympy
```

```python
# a symbolic variable for x, and 
# a numerical array with specific values of x
sym_x = sympy.Symbol("x")
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
```

```python
def sin_expansion(sym_x, x, n):
    """
    Evaluate the nth order Taylor. series expansion
    of sin(x) for the numerical values in the array x.
    """
    return sympy.lambdify(sym_x, sympy.sin(sym_x).series(n=n).removeO(), 'numpy')(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), linewidth=4, color="red", label='exact')

colors = ["blue", "black"]
linestyles = [':', '-.', '--']
for idx, n in enumerate(range(2, 11, 2)):

    ax.set_ylim(-1.1, 1.1)
    ax.set_xlim(-1.5*np.pi, 1.5*np.pi)
    ax.plot(x, sin_expansion(sym_x, x, n), 
            color=colors[idx // 3],
            linestyle=linestyles[idx % 3], 
            linewidth=3,
            label=f"order {n} approx.")

# place a legend outsize of the Axes
ax.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.0)
# make room for the legend to the right of the Axes
fig.subplots_adjust(right=.75)
```
