```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
ticker = 'AAPL'
start = '2020-01-01'
end = '2020-12-31'
df = yf.Ticker(ticker).history(start=start, end=end)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
df.Close.plot(ax=axes[0])
df.Volume.plot(ax=axes[1]);
```
