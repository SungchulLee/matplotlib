```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 5)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(
    x, y,
    color='gray',
    linestyle='-',
    linewidth=4,
    marker='p',
    markersize=15,
    markeredgecolor='red',
    markeredgewidth=2,
    markerfacecolor='blue'
);
```
