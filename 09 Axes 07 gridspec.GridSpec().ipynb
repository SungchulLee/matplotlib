```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,3)
y1 = x ** 2
y2 = np.sqrt(x)
```

```python
fig = plt.figure()
gs = gridspec.GridSpec(2, 3, height_ratios=[2,1], width_ratios=[1,2,1])
for i, g in enumerate(gs):
    ax = fig.add_subplot(g)
    if i == 0:
        ax.plot(x,y1)
    elif i == 1: 
        ax.plot(y1,x)
    elif i == 2:
        ax.plot(x,y1)
    elif i == 3:
        ax.plot(x,y2)
    elif i == 4:
        ax.plot(y2,x)
    elif i == 5:
        ax.plot(x,y2)
fig.tight_layout();
```
