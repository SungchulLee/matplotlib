```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x) * np.exp(-x**2/20)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, lw=2)
ax.set_yticks([-1, 0, 1])
ax.set_xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi]) 
ax.set_xticklabels(['$-2\pi$', '$-\pi$', 0, '$\pi$', '$2\pi$']);
```
