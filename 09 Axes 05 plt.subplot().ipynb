# MATLAB-like API

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,3)
y1 = x ** 2
y2 = np.sqrt(x)
```

```python
#plt.figure()

if 0:
    plt.plot(x,y1)
elif 1:
    plt.subplot(2,3,1)
    plt.plot(x,y1)
    plt.subplot(2,3,2)
    plt.plot(y1,x)
    plt.subplot(2,3,3)
    plt.plot(x,y1)
    plt.subplot(2,3,4)
    plt.plot(x,y2)
    plt.subplot(2,3,5)
    plt.plot(y2,x)
    plt.subplot(2,3,6)
    plt.plot(x,y2)
```
