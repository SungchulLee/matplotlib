```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
y = x + x**2
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, color='g', label="y(x)=x+x**2")
ax.legend()

ax.set_title('$y=x+x^2$');
```
