```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(337)
```

```python
x1 = np.random.normal(0, 1, (10000,))
x2 = np.random.normal(0, 2, (10000,))
```

```python
fig, ax = plt.subplots()
ax.hist(x1, bins=30, density=True, color='y', label='Type 1')
ax.hist(x2, bins=30, density=True, alpha=0.5, label='Type 2')
ax.legend(loc='upper right', shadow=True, fontsize='large')  
ax.set_xlabel("Sugar Level")
ax.set_ylabel("Number Of Patients")
ax.set_title("Blood Sugar Chart")
```
