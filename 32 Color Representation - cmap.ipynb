```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(type(x_train))
print(type(y_train))
print(type(x_test))
print(type(y_test))
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
print(x_train.dtype)
print(y_train.dtype)
print(x_test.dtype)
print(y_test.dtype)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=10, figsize=(10,2))
for i in range(2):
    for j in range(10):
        axes[i, j].imshow(x_train[i*10+j], cmap=plt.cm.gray) 
        #axes[i, j].imshow(x_train[i*10+j], cmap='prism') 
        axes[i, j].set_title('Label {}'.format(y_train[i*10+j]))
        axes[i, j].axis('off')
plt.tight_layout();
```
