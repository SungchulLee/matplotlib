```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
if 0:
    plt.plot([1, 2, 4, 9, 5, 3])
elif 1:
    plt.plot([0, 1, 2, 3, 4, 5], [1, 2, 4, 9, 5, 3])
    
plt.show()
```

```python
if 1:
    plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
elif 1:
    plt.plot([-3, -2, 5, 0], [1, 6, 4, 3], 'o')
plt.show()
```

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
if 0:
    plt.axis([-4, 6, 0, 7])
elif 1:
    plt.xlim([-4, 6])
    plt.ylim([0, 7])    
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Square function $y=x^2$")
plt.xlabel("x")
plt.ylabel("$y = x^2$")
plt.grid(True)
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
if 0:
    plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
elif 1:
    plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
    plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
if 0:
    plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
elif 1:
    plt.plot(x, x, 'g--')
    plt.plot(x, x**2, 'r:')
    plt.plot(x, x**3, 'b*')
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
if 0:
    line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
    line1.set_linewidth(3.0)
    line1.set_dash_capstyle("round")
    line3.set_alpha(0.2)
elif 0:
    line1 = plt.plot(x, x, 'g--')
    line2 = plt.plot(x, x**2, 'r:')
    line3 = plt.plot(x, x**3, 'b^')
    line1[0].set_linewidth(3.0)
    line1[0].set_dash_capstyle("round")
    line3[0].set_alpha(0.2)
elif 1:
    plt.plot(x, x, 'g--', linewidth=3.0)
    plt.plot(x, x**2, 'r:', dash_capstyle="round")
    plt.plot(x, x**3, 'b^', alpha=0.2)
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x**2)
plt.savefig("my_square_function.png", transparent=True)
```

```python
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1)      # back to figure 1, current subplot is 212 (bottom)
plt.plot(x, -x**3, "r:")

plt.show()
```
