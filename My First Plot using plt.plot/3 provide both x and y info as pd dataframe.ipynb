# Easy Fix

```python
import matplotlib.pyplot as plt
import pandas as pd

x = pd.DataFrame([[1,2,3]])
y = pd.DataFrame([[1,5,2]])
 
plt.plot(x.iloc[0,:], y.iloc[0,:])

plt.show()
```
