```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = np.linspace(0, 5, 100)
a = 4 * np.pi * r ** 2  # area
v = (4 * np.pi / 3) * r ** 3  # volume
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.set_title("surface area and volume of a sphere", fontsize=16)
ax.set_xlabel("radius (m)", fontsize=16)

ax.plot(r, a, lw=2, color="blue")
ax.set_ylabel("surface area ($m^2$)", fontsize=16, color="blue")
for label in ax.get_yticklabels():
    label.set_color("blue")

ax2 = ax.twinx()
ax2.plot(r, v, lw=2, color="red")
ax2.set_ylabel("volume ($m^3$)", fontsize=16, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
