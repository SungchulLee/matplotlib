```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Generate MA(2)

$$y_t = 50 + 0.4\varepsilon_{t-1} + 0.3\varepsilon_{t-2} + \varepsilon_t$$
where
$$\varepsilon_t \sim N(0,1)$$

```python
error = np.random.normal(size=(400,))
```

```python
index = pd.date_range(start='2019-09-01', end='2020-01-01', freq='D')
```

```python
mu = 50
data = []
for t in range(2, len(index)+2):
    data.append(mu + 0.4*error[t-1] + 0.3*error[t-2] + error[t])
```

```python
s = pd.Series(data, index)
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax.plot(s)
ax.axhline(mu, linestyle='--', color='grey')
for label in ax.get_xticklabels():
    label.set_ha("right") # ha = horizontal alignment
    label.set_rotation(45);
```
